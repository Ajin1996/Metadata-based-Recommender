```python
def sanitize(x):
    if isinstance(x, list):
        #Strip spaces and convert to lowercase
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
for feature in ['title', 'cast', 'crew','director','year', 'keywords', 'genres']:
    filter_df[feature] = filter_df[feature].apply(sanitize)
```

## Create a soup of desired metadata 

```python
def soup(x):
    return ' '.join(x['title']) + ' ' + ' '.join (x['cast']) + ' ' + ' '.join(x['crew'])+ ' ' + ' '.join(x['director'])+ ' ' + ' '.join(x['genres'])+ ' ' + ' '.join(x['year'])+ ' ' + ' '.join(x['keywords'])
```

```python
filter_df['soup'] = filter_df.apply(soup, axis=1)
```

```python
filter_df.iloc[0]['soup']
```

## Create vectors with CountVectorizer 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(filter_df['soup'])
```

### pairwise similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

## Build the Metadata-based Recommender

```python
# Reset index of your df and construct reverse mapping again
filter_df = filter_df.reset_index()
indices = pd.Series(df.index, index=df['title'])
```

```python
# Function that takes in movie title as input and gives recommendations 
def content_recommender(title, cosine_sim=cosine_sim, df=filter_df, indices=indices):
    # Obtain the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies. Ignore the first movie.
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title'].iloc[movie_indices]
```
